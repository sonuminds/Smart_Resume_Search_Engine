```python
def create_resume_text(resume):
    return f"""
    Name: {resume['name']}
    Role: {resume['role']}
    Skills: {', '.join(resume['skills'])}
    Experience: {resume['experience']}
    Education: {resume['education']}
    Summary: {resume['summary']}
    """
```

```python
sample_text = create_resume_text(resumes[0])

print(sample_text)
```

```python
import google.generativeai as genai

def get_embedding(text):
    response = genai.embed_content(
        model="models/gemini-embedding-001",
        content=text
    )

    return response["embedding"]
```

```python
test_embedding = get_embedding(create_resume_text(resumes[0]))

print("Embedding length:", len(test_embedding))
```

```python
import numpy as np
import json

def build_embeddings(resumes):

    embeddings = []
    metadata = []

    for resume in resumes:

        text = create_resume_text(resume)

        vector = get_embedding(text)

        embeddings.append(vector)

        metadata.append({
            "name": resume["name"],
            "role": resume["role"],
            "skills": resume["skills"],
            "summary": resume["summary"]
        })

        print(f"Embedded: {resume['name']}")

    embeddings = np.array(embeddings).astype("float32")

    np.save("resumes.npy", embeddings)

    with open("resumes_meta.json", "w") as f:
        json.dump(metadata, f, indent=4)

    print("\nSaved Successfully!")
    print("resumes.npy")
    print("resumes_meta.json")
```

```python
import numpy as np
import json

def cosine_similarity(query_vector, resume_vectors):
    query_vector = query_vector / np.linalg.norm(query_vector)
    resume_vectors = resume_vectors / np.linalg.norm(resume_vectors, axis=1, keepdims=True)
    return np.dot(resume_vectors, query_vector)

def search_candidates_numpy(jd, top_k=5):
    vectors = np.load("resumes.npy")

    with open("resumes_meta.json", "r") as f:
        metadata = json.load(f)

    jd_vector = np.array(get_embedding(jd)).astype("float32")

    scores = cosine_similarity(jd_vector, vectors)

    top_indices = np.argsort(scores)[::-1][:top_k]

    results = []

    for index in top_indices:
        if scores[index] >= 0.55:
            candidate = metadata[index].copy()
            candidate["score"] = round(float(scores[index]), 4)
            results.append(candidate)

    if len(results) == 0:
        return "No strong matches found"

    return results
```

```python
tech_jd = """
Looking for a Machine Learning student with Python, pandas, scikit-learn,
data analysis, classification models, and experience with predictive analytics.
"""

numpy_results = search_candidates_numpy(tech_jd, top_k=5)

for i, candidate in enumerate(numpy_results, start=1):
    print(f"\nRank {i}")
    print("Name:", candidate["name"])
    print("Role:", candidate["role"])
    print("Skills:", ", ".join(candidate["skills"]))
    print("Score:", candidate["score"])
```

```python
import faiss
import numpy as np
import json

def search_candidates_faiss(jd, top_k=5):
    vectors = np.load("resumes.npy").astype("float32")

    with open("resumes_meta.json", "r") as f:
        metadata = json.load(f)

    faiss.normalize_L2(vectors)

    dimension = vectors.shape[1]

    index = faiss.IndexFlatIP(dimension)
    index.add(vectors)

    jd_vector = np.array([get_embedding(jd)]).astype("float32")
    faiss.normalize_L2(jd_vector)

    scores, indices = index.search(jd_vector, top_k)

    results = []

    for score, index_value in zip(scores[0], indices[0]):
        if score >= 0.55:
            candidate = metadata[index_value].copy()
            candidate["score"] = round(float(score), 4)
            results.append(candidate)

    if len(results) == 0:
        return "No strong matches found"

    return results
```

```python
faiss_results = search_candidates_faiss(tech_jd, top_k=5)

for i, candidate in enumerate(faiss_results, start=1):
    print(f"\nRank {i}")
    print("Name:", candidate["name"])
    print("Role:", candidate["role"])
    print("Skills:", ", ".join(candidate["skills"]))
    print("Score:", candidate["score"])
```

```python
non_tech_jd = """
Looking for an HR Executive with experience in recruitment,
onboarding, payroll management, and employee engagement.
"""

results = search_candidates_faiss(non_tech_jd)

for i, candidate in enumerate(results, start=1):
    print(f"\nRank {i}")
    print(candidate["name"])
    print(candidate["role"])
    print(candidate["score"])
```
